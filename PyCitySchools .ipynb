```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# File to Load 
school_data_to_load = ("../Documents/Pandas/schools_complete.csv")
student_data_to_load = ("../Documents/Pandas/students_complete.csv")

# Read School and Student Data File and store into Pandas Data Frames

school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

```python
#Calculate the total number of schools
got_school_name = school_data_complete["school_name"].value_counts()
Total_Schools1 = got_school_name.count()
Total_Schools1
```

```python
#Calculate the total number of students
got_students_name = school_data_complete['student_name'].value_counts()
Total_Students1 = got_students_name.sum()
Total_Students1
```

```python
#Calculate the total budget
Final_Total_Budget = school_data['budget'].sum()
Final_Total_Budget
```

```python
#Calculate the average math score 
Average_math_scores = school_data_complete['math_score'].mean()
Average_math_scores
```

```python
#Calculate the average reading score
Average_Reading_score = school_data_complete['reading_score'].mean()
Average_Reading_score
```

```python
#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

Overall_Passing_Rate = (Average_Reading_score + Average_math_scores)/2
Overall_Passing_Rate
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)

Avg_Math_score = school_data_complete.loc[(school_data_complete["math_score"] >= 70)]
Numberstudents70 = Avg_Math_score['student_name'].count()
Number170 = (Numberstudents70 *100)/ Total_Students1
Number170
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater
Avg_Read_score = school_data_complete.loc[(school_data_complete["reading_score"] >= 70)]
ReadNumberstudents70 = Avg_Read_score['student_name'].count()
ReadNumber170 = (ReadNumberstudents70 *100)/ Total_Students1
ReadNumber170
```

```python
District_Summary = pd.DataFrame({
                                "Total Schools": [Total_Schools1],
                                "Total Students":[Total_Students1],
                                "Total Budget": [Final_Total_Budget],
                                "Average Math Score": [Average_math_scores],
                                "Average Reading Score": [Average_Reading_score], 
                                "% Passing Math": [Number170],
                                "% Passing Reading":[ReadNumber170],
                                 "Overall Passing Rate":[Overall_Passing_Rate]})

#format cells
District_Summary.style.format({"Total Students":"{:,}",
                        "Total Budget": "${:,.2f}", 
                       "Average Reading Score": "{:.6f}", 
                       "Average Math Score": "{:.6f}", 
                       "% Passing Math": "{:.6f}", 
                       "% Passing Reading": "{:.6f}", 
                       "Overall Passing Rate": "{:.6f}"})
```

# Scores by School Spending
 
Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
Average Math Score
Average Reading Score
% Passing Math
% Passing Reading
Overall Passing Rate (Average of the above two)

```python
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
school_data_complete['spending_bins'] = pd.cut(school_data_complete['budget']/school_data_complete['size'], spending_bins,
                                                     labels = group_names)
#Group by Spending
Spending = school_data_complete.groupby('spending_bins')

Avg_Math_Spending = Spending['math_score'].mean()
Avg_Read_Spending = Spending['reading_score'].mean()
Pass_Math_Spending1 = school_data_complete[school_data_complete
                                            ['math_score'] >= 70].groupby('spending_bins')['Student ID'].count()/Spending['Student ID'].count()
Pass_Math_Spending = (Pass_Math_Spending1 *100)
Pass_Read_Spending1 = school_data_complete[school_data_complete
                                            ['reading_score'] >= 70].groupby('spending_bins')['Student ID'].count()/Spending['Student ID'].count()
Pass_Read_Spending = (Pass_Read_Spending1 * 100)
Overall_Score_Spending= (Pass_Math_Spending + Pass_Read_Spending)/2
            
# New Dataframe           
Scores_by_Spend = pd.DataFrame({
    "Average Math Score": Avg_Math_Spending,
    "Average Reading Score": Avg_Read_Spending,
    '% Passing Math': Pass_Math_Spending,
    '% Passing Reading': Pass_Read_Spending,
    "% Overall Passing Rate": Overall_Score_Spending})
        

Scores_by_Spend.index.name = "Spending Ranges (Per Student)"
Scores_by_Spend = Scores_by_Spend.reindex(group_names)

#formating
Scores_by_Spend.style.format({'Average Math Score': '{:.6f}', 
                              'Average Reading Score': '{:.6f}', 
                              '% Passing Math': '{:.6f}', 
                              '% Passing Reading':'{:.6f}', 
                              'Overall Passing Rate': '{:.6f}'})
```
